```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

import warnings
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-means-project-tutorial/main/housing.csv")
df
```

```python
df=df[["Latitude","Longitude","MedInc"]]
df
```

```python
# Dividimos la muestra en train y test al 80%
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, random_state = 42, train_size = 0.80)
df_test.shape, df_train.shape
```

```python
#num de clusters 6
# Entrenar el modelo
model = KMeans(n_clusters = 6, random_state = 42)
model.fit(df_train)
```

```python
labels = model.labels_
df_train["cluster"]=labels
df_train['cluster'] = df_train['cluster'].astype('category')
df_train
```

```python
# Graficar la distribución de los datos
import matplotlib.pyplot as plt

plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, c='blue')
plt.title("Distribución Geográfica")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Seleccionar las columnas Latitude y Longitude para clustering
geo_data = df[['Latitude', 'Longitude']]

# Aplicar K-Means
kmeans = KMeans(n_clusters=4, random_state=42)  # Cambia n_clusters según el análisis
df['Cluster'] = kmeans.fit_predict(geo_data)

# Graficar los clusters
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', alpha=0.7)
plt.title("Clustering Geográfico")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# Graficar Latitud vs Longitud con colores por cluster
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x='Longitude', y='Latitude',
    hue='cluster', data=df_train,
    palette='tab10', s=100
)
plt.title('Clusters según Latitud y Longitud', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='Cluster')
plt.show()
```

se puede ver una clara separacion en dos grupos. En la parte inferior derecha estan ubicados los cluster 1,2 y 3 mientras que en la parte superior izquerda los cluster 0,4 y 5

```python
# Graficar Latitud vs Longitud con colores por cluster
#camnbio en el tamaño de los puntos "s"
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x='Longitude', y='Latitude',
    hue='cluster', data=df_train,
    palette='tab10', s=50
)
plt.title('Clusters según Latitud y Longitud', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='Cluster')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib import cm

# Crear un colormap con colores según los clusters
n_clusters = df_train['cluster'].nunique()  # Número de clusters
cmap = ListedColormap(cm.get_cmap('tab10').colors[:n_clusters])

# Configurar los subplots
fig = plt.figure(figsize=(16, 6))  # Ajusta el tamaño (ancho, alto)

# Primer gráfico: Longitude vs Latitude
ax1 = fig.add_subplot(121, projection='3d')  # 1 fila, 2 columnas, posición 1
scatter1 = ax1.scatter(
    df_train['Longitude'], df_train['Latitude'], df_train['MedInc'],
    c=df_train['cluster'].cat.codes,  # Convertir clusters a códigos numéricos
    cmap=cmap, s=100
)
ax1.set_title('Clusters: Longitude vs Latitude', fontsize=16)
ax1.set_xlabel('Longitude', fontsize=12)
ax1.set_ylabel('Latitude', fontsize=12)
ax1.set_zlabel('MedInc', fontsize=12)

# Barra de colores para el primer gráfico
cbar1 = fig.colorbar(scatter1, ax=ax1, pad=0.1)
cbar1.set_label('Cluster', fontsize=12)

# Segundo gráfico: Latitude vs Longitude
ax2 = fig.add_subplot(122, projection='3d')  # 1 fila, 2 columnas, posición 2
scatter2 = ax2.scatter(
    df_train['Latitude'], df_train['Longitude'], df_train['MedInc'],
    c=df_train['cluster'].cat.codes,
    cmap=cmap, s=100
)
ax2.set_title('Clusters: Latitude vs Longitude', fontsize=16)
ax2.set_xlabel('Latitude', fontsize=12)
ax2.set_ylabel('Longitude', fontsize=12)
ax2.set_zlabel('MedInc', fontsize=12)

# Barra de colores para el segundo gráfico
cbar2 = fig.colorbar(scatter2, ax=ax2, pad=0.1)
cbar2.set_label('Cluster', fontsize=12)

# Ajustar el espacio entre gráficos y mostrar
plt.tight_layout()
plt.show()
```

```python
X_train=df_train.drop("cluster", axis=1)
y_train=df_train["cluster"]

X_test=df_test.drop("cluster", axis=1)
y_test=df_test["cluster"]
```

```python
from sklearn.ensemble import RandomForestClassifier  # Random Forest
# Inicializar los modelos
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)  # Random Forest
```

```python
# entrenamiento
random_forest.fit(X_train, y_train)
```

```python
# predicción
random_forest_y_pred_test = random_forest.predict(X_test)

random_forest_y_pred_train = random_forest.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
# metricas
random_forest_accuracy_test = accuracy_score(y_test, random_forest_y_pred_test)
random_forest_accuracy_train = accuracy_score(y_train, random_forest_y_pred_train)
```

```python
# Imprimir métricas
print(f"Precisión del modelo Random Forest en el conjunto de entrenamiento: {random_forest_accuracy_train:.2f}")
print(f"Precisión del modelo Random Forest en el conjunto de prueba: {random_forest_accuracy_test:.2f}")
```
